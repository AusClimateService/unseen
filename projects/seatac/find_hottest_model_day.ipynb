```python
ds = fileio.open_file(infile)
```

```python
ds
```

```python
argmax = ds['tasmax'].argmax(dim=['ensemble', 'init_date', 'lead_time'])
```

```python
ensemble_idx = int(argmax['ensemble'].values)
print(ensemble_idx)
```

```python
init_date_idx = int(argmax['init_date'].values)
print(init_date_idx)
```

```python
lead_time_idx = int(argmax['lead_time'].values)
print(lead_time_idx)
```

```python
max_ds = ds['tasmax'].isel({'ensemble': ensemble_idx, 'init_date': init_date_idx, 'lead_time': lead_time_idx})
```

```python
max_ds
```
