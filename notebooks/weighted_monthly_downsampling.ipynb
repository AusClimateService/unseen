## Seasonal

```python
months = ds.indexes['time'].month
months
```

```python
days_in_month = ds['time'].dt.days_in_month
```

```python
days_in_month
```

```python
start_month = months[0]
start_month
```

```python
end_month = months[-1]
end_month
```

```python
rolling_sum = days_in_month.rolling({'time': 3}, min_periods=1).sum()
rolling_sum
```

```python
if start_month in [2, 5, 8, 11]:
    start_index = 0
elif start_month in [1, 4, 7, 10]:
    start_index = 1
elif start_month in [12, 3, 6, 9]:
    start_index = 2

days_per_season = rolling_sum[start_index::3]
days_per_season
```

```python
days_in_season = np.repeat(days_per_season.values[1:], 3)
days_in_season
```

```python
if start_month in [2, 5, 8, 11]:
    days_in_season = np.insert(days_in_season, 0, rolling_sum[0])
elif start_month in [1, 4, 7, 10]:
    days_in_season = np.insert(days_in_season, 0, rolling_sum[0:2])
```

```python
if end_month in [1, 4, 7, 10]:
    final_season = days_in_month[-1] + days_in_month[-2]
    days_in_season = np.append(days_in_season, [final_season, final_season])
elif end_month in [12, 3, 6, 9]:
    days_in_season = np.append(days_in_season, days_in_month[-1])
```
