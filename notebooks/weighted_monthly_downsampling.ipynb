```python
import sys
sys.path.append('../unseen')

import xarray as xr
import numpy as np

import myfuncs
```

```python
infile = '/g/data/xv83/ds0092/CAFE/forecasts/f6/WIP/c5-d60-pX-f6-19931101/ZARR/atmos_isobaric_month.zarr.zip'
```

```python
cafe_kwargs = {'metadata_file': '../config/dataset_cafe.yml',
               'no_leap_days': True,
               'region': 'TAS-POINT',
               'variables': ['pr'],
               'units' : {'pr': 'mm/day'},
              }
```

```python
ds = myfuncs.open_file(infile, **cafe_kwargs)
ds
```

```python
days_in_month = ds.time.dt.days_in_month
```

```python
days_in_year = days_in_month.groupby('time.year').sum().values
```

```python
days_in_year
```

```python
years = ds.indexes['time'].year
unique_years = np.unique(years)
```

```python
unique_years
```

```python
days_in_unique_years = {unique_years[i]: days_in_year[i] for i in range(len(unique_years))}
```

```python
days_in_unique_years
```

```python
days_in_year = np.array([days_in_unique_years[yr] for yr in years])
```

```python
days_in_year
```

```python
annual_fraction = days_in_month / days_in_year
```
